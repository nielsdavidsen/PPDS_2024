```python
import scipy.optimize as opt
import numpy as np
import matplotlib.pyplot as plt
```

```python
#task at hand: minimization of a simple trig function

def neg_f(x):
    return -(np.cos(x) + np.sin(2*x))
```

```python
#quick vizualization of the function
linsoace = np.linspace(-2*np.pi, 2*np.pi, 100)
y = neg_f(linsoace)

plt.plot(linsoace,y)
```

```python
#for minimizing we use our opt with the powell method. 
# The Powell method is a conjugate direction method. 
# It performs sequential one-dimensional minimizations along each vector of the directions set 
# (which is updated at each iteration of the main minimization loop). 
# This method is useful for functions that are not smooth or have discontinuities. 


minimum = opt.minimize(neg_f, x0=0.0, method='powell')
print(minimum.x)

#vizualization of minimum
minimum_y = neg_f(minimum.x)
plt.plot(linsoace, y)
plt.plot(minimum.x, minimum_y, 'ro')
```

```python
#it has simply found one mimimum, which is the closest to my initial guess of 0.0
```

```python
#Now let's integrate the function using scipy. Our result will be a numberical integration ofc
import scipy.integrate as integ

def f(x):
    return np.cos(x) + np.sin(2*x)

#integrate from -pi to pi
res, err = integ.quad(f, -np.pi, np.pi)

print(res)
```

```python
#Next we will interpolate between a few points.

from scipy.interpolate import CubicSpline

#the points we will interpolate
x = np.array([0, 1, 2, 3, 4])
y = np.array([12,14,22,28,20])

#the spline (a "spline" refers to a piecewise polynomial function that can provide a smooth approximation to a set of data points. 
# Splines are particularly useful for interpolation because they can create smooth curves that pass through 
# or near the given data points.)
cs = CubicSpline(x,y,bc_type='natural')

# Evaluate spline at new points
x2 = np.arange(0, 4.2, 0.1)
y2 = cs(x2)

plt.plot(x,y, '.', markersize=10)
plt.plot(x2,y2)
```
